```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import os
env_name = 'Pendulum-v0'
path = './csv_log/'+env_name
file_list = os.listdir(path)
```

```python
data_list = []
for file in file_list:
    data_list.append(pd.read_csv(path+'/'+file))
for i, data in enumerate(data_list):
    data_list[i] = pd.DataFrame([data_list[i]['step'],data_list[i]['value']]).T
    data_list[i] = data_list[i].rename(columns={'step':'#Step','value':'Return'})
    #data_list.append(data)
```

```python
final = pd.concat([data for data in data_list],axis=0)
```

```python
g = sns.relplot('#Step','Return', kind="line",ci='sd', data=final)
g.fig.suptitle('TRPO with Different Random Seed',x=0.6,y=1.05)
plt.show()
```
