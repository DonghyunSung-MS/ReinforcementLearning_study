```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
import math
```

```python
#normal distribution
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.plot(x, stats.norm.pdf(x, mu, 2*sigma))
plt.show()
```

```python
#entropy
print(stats.entropy([1/2,1/2]))
print(stats.entropy([0.9,0,0.1]))
```
